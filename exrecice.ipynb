# importing libraries

```python
import requests as rq
from bs4 import BeautifulSoup as soup
```

```python
my_url="https://pokemondb.net/pokedex/all"
```

# opening up connection , grabbing 

```python
req = rq.get(my_url)
status= req.status_code
print(status)
```

```python
encoding = req.encoding
html_doc = req.text
```

```python
Soup =soup(html_doc,"html.parser")
data=Soup.find_all("span",{"class":"infocard-cell-img"})
data1=Soup.find_all("span",{"class":"infocard-cell-data"})
data2=Soup.find_all("td",{"class":"cell-name"})
print ("Liste des pokémons :")
print ("id | name")
for item1,item2 in zip(data1,data2):
    print(item1.contents[0]+" "+item2.contents[0].text)


    
    
       
 
```
